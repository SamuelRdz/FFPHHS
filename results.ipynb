```python
import pandas as pd

df = pd.read_csv('final_df.csv')
df
```

```python
df.mean()
```

```python
df['best_LDEG'] = (df['LDEG'] <= df['GDEG']) & (df['LDEG'] <= df['Oracle']) & (df['LDEG'] <= df['PBC4cip'])
df['best_GDEG'] = (df['GDEG'] <= df['LDEG']) & (df['GDEG'] <= df['Oracle']) & (df['GDEG'] <= df['PBC4cip'])
df['best_Oracle'] = (df['Oracle'] <= df['GDEG']) & (
    df['Oracle'] <= df['LDEG']) & (df['Oracle'] <= df['PBC4cip'])
df['best_PBC4cip'] = (df['PBC4cip'] <= df['GDEG']) & (
    df['PBC4cip'] <= df['Oracle']) & (df['PBC4cip'] <= df['LDEG'])

```

```python
df[df['best_PBC4cip'] == True].count()
df.count()
```

```python
df['best_LDEG_vs_GDEG'] = (df['LDEG'] <= df['GDEG'])
df['best_GDEG_vs_LDEG'] = (df['GDEG'] <= df['LDEG'])
df['best_LDEG_vs_GDEG_unique'] = (df['LDEG'] < df['GDEG'])
df['best_GDEG_vs_LDEG_unique'] = (df['GDEG'] < df['LDEG'])
df[df['best_GDEG_vs_LDEG_unique'] == True].count()
```

```python
df['best_LDEG_vs_GDEG_vs_PBC4cip'] = (df['LDEG'] <= df['GDEG']) & (df['LDEG'] <= df['PBC4cip'])
df['best_GDEG_vs_LDEG_vs_PBC4cip'] = (df['GDEG'] <= df['LDEG']) & (df['GDEG'] <= df['PBC4cip'])
df['best_PBC4cip_vs_LDEG_vs_GDEG'] = (df['PBC4cip'] <= df['LDEG']) & (df['PBC4cip'] <= df['GDEG'])
df['best_LDEG_vs_GDEG_vs_PBC4cip_unique'] = (df['LDEG'] < df['GDEG']) & (df['LDEG'] < df['PBC4cip'])
df['best_GDEG_vs_LDEG_vs_PBC4cip_unique'] = (df['GDEG'] < df['LDEG']) & (df['GDEG'] < df['PBC4cip'])
df['best_PBC4cip_vs_LDEG_vs_GDEG_unique'] = (df['PBC4cip'] < df['LDEG']) & (df['PBC4cip'] < df['GDEG'])
df[df['best_PBC4cip_vs_LDEG_vs_GDEG_unique'] == True].count()
```

```python
df['best_PBC_vs_Oracle'] = (df['PBC4cip'] <= df['Oracle'])
df['best_Oracle_vs_PBC'] = (df['Oracle'] <= df['PBC4cip'])
df['best_PBC_vs_Oracle_unique'] = (df['PBC4cip'] < df['Oracle'])
df['best_Oracle_vs_PBC_unique'] = (df['Oracle'] < df['PBC4cip'])
df[df['best_Oracle_vs_PBC_unique'] == True].count()
```
